```python
import pandas as pd
import numpy as np

data = pd.read_csv("habitat_data.csv", usecols = ['User', 'World', 'Block_Type'])
builders = pd.read_csv('builders.csv')
builders.head(5)
```

```python
data.World.unique()
```

```python
data.User.unique()
```

```python
blocks = data.Block_Type.unique()
blocks
```

```python
r = 0
block_counts = pd.DataFrame(columns = blocks)
for index, row in builders.iterrows():
    world = data[data['World'] == row['world']]
    if pd.isna(row['builder2']):
        if pd.isna(row['builder3']):
            base = world[(world['User'] == row['builder1'])]
        else:
            base = world[(world['User'] == row['builder1']) | (world['User'] == row['builder2'])]
    else:
        base = world[(world['User'] == row['builder1']) | (world['User'] == row['builder2']) | (world['User'] == row['builder3'])]
    counts = base.Block_Type.value_counts()
    for i, v in counts.items():
        block_counts.at[r,i] = v
    print(base.User.unique())
    r += 1
block_counts.head(5)
```

```python
block_counts.dropna(axis=1, how='all', inplace = True)
block_counts.fillna(0, inplace=True)
block_counts.to_csv('block_counts.csv',index = False)
block_counts
```

```python
import scipy
from scipy import stats

normalized_counts = block_counts.copy()
standardized_counts = block_counts.copy()
for name, values in block_counts.iteritems():
    normalized_counts[name] = (normalized_counts[name] - normalized_counts[name].min()) / (normalized_counts[name].max() - normalized_counts[name].min()) 
    standardized_counts[name] = stats.zscore(standardized_counts[name])
```

```python
normalized_counts.fillna(0, inplace=True)
normalized_counts.to_csv('block_counts_normalized.csv',index = False)
normalized_counts
```

```python
standardized_counts.fillna(0, inplace=True)
standardized_counts.to_csv('block_counts_standardized.csv',index = False)
standardized_counts
```
