```python
import sqlite3
import pandas as pd

db_file = "DBs/core protect summer 2023 bak.db"
con = sqlite3.connect(db_file)
cur = con.cursor()
world = "goldcrown"
```

```python
teammates = pd.read_csv('builders.csv')
teammates.head(5)
```

```python
num_teammates = teammates.count(axis=1)
num_teammates
```

```python
worlds = pd.read_sql_query("SELECT * from co_world", con)
worldID = worlds.loc[worlds['world'] == world]['id'].values[0]
worldID
```

```python
blocks = pd.read_sql_query("SELECT * from co_block WHERE wid=" + str(worldID) + ";", con)
blocks.user.unique()
```

```python
users = pd.read_sql_query("SELECT * from co_user", con)
users.user.unique()
```

```python
users
```

```python
materials = pd.read_sql_query("SELECT * from co_material_map", con)
materials.head(5)
```

```python
blockTypes = materials.material.unique()
blockTypes
```

```python
#Calculate count of blocks for each group and set value for each block type
block_counts = pd.DataFrame(columns = blockTypes)
groups = teammates.loc[teammates['world'] == world]
for index, row in groups.iterrows():
    group_size = num_teammates.iloc[[index]][index]-2
    print(row['id'])
    if group_size == 1:
        b1ID = users.loc[users['user'] == row['builder1']]['id'].values[0]
        base = blocks[(blocks['user'] == b1ID)]
    elif group_size == 2:    
        b1ID = users.loc[users['user'] == row['builder1']]['id'].values[0]
        b2ID = users.loc[users['user'] == row['builder2']]['id'].values[0]
        base = blocks[(blocks['user'] == b1ID) | (blocks['user'] == b2ID)]
    elif group_size == 3:
        b1ID = users.loc[users['user'] == row['builder1']]['id'].values[0]
        b2ID = users.loc[users['user'] == row['builder2']]['id'].values[0]
        b3ID = users.loc[users['user'] == row['builder3']]['id'].values[0]
        base = blocks[(blocks['user'] == b1ID) | (blocks['user'] == b2ID) | (blocks['user'] == b3ID)]
    elif group_size == 4:
        b1ID = users.loc[users['user'] == row['builder1']]['id'].values[0]
        b2ID = users.loc[users['user'] == row['builder2']]['id'].values[0]
        b3ID = users.loc[users['user'] == row['builder3']]['id'].values[0]
        b4ID = users.loc[users['user'] == row['builder4']]['id'].values[0]
        base = blocks[(blocks['user'] == b1ID) | (blocks['user'] == b2ID) | (blocks['user'] == b3ID) | (blocks['user'] == b4ID)]
    elif group_size == 5:
        b1ID = users.loc[users['user'] == row['builder1']]['id'].values[0]
        b2ID = users.loc[users['user'] == row['builder2']]['id'].values[0]
        b3ID = users.loc[users['user'] == row['builder3']]['id'].values[0]
        b4ID = users.loc[users['user'] == row['builder4']]['id'].values[0]
        b5ID = users.loc[users['user'] == row['builder5']]['id'].values[0]
        base = blocks[(blocks['user'] == b1ID) | (blocks['user'] == b2ID) | (blocks['user'] == b3ID) | (blocks['user'] == b4ID) | (blocks['user'] == b5ID)]
    counts = base.type.value_counts()
    for i, v in counts.items():
        material = materials.loc[materials['id'] == i]
        block_counts.at[row['id'],material['material']] = v
block_counts.head(5)
```

```python
block_dataset = pd.read_csv("block_counts.csv")
block_counts = block_counts.reindex(block_dataset.columns,axis=1)
block_counts
```

```python
block_counts.to_csv(world+"_block_counts.csv")
```
